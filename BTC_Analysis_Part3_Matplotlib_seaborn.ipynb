# ðŸ“Š BTC Dataset Visualization using Matplotlib and Seaborn

In this notebook, we will visualize various aspects of the BTC dataset using Matplotlib and Seaborn.  
Make sure that you've already loaded the dataset into the DataFrame `df` and defined `close_prices` (if available).


```python
# Import the necessary libraries for visualization
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("btc_15m_data_2018_to_2025.csv")



# Ensure that plots are displayed inline (if using a notebook)
%matplotlib inline

# Optional: set the Seaborn style
sns.set(style="whitegrid")
```

```python
# Example 1: Line Plot of the 'close' prices (if time index is available)

# If you have converted your 'date' or 'timestamp' column and set it as index:
if df.index.name is not None:
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['close'], label="Close Price")
    plt.title("BTC Close Price Over Time")
    plt.xlabel("Date")
    plt.ylabel("Close Price")
    plt.legend()
    plt.show()
else:
    print("No datetime index found. Make sure you've converted your date column and set it as index.")
```

```python
# Example 2: Histogram of the Close Prices using Matplotlib
if 'close' in df.columns:
    plt.figure(figsize=(10, 5))
    plt.hist(df['close'], bins=50, color='skyblue', edgecolor='black')
    plt.title("Distribution of BTC Close Prices")
    plt.xlabel("Close Price")
    plt.ylabel("Frequency")
    plt.show()
else:
    print("Column 'close' not found.")
```

```python
# Example 3: Seaborn Distribution Plot (KDE) for the Close Prices
if 'close' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df['close'], kde=True, bins=50, color='teal')
    plt.title("Seaborn Distribution and KDE of BTC Close Prices")
    plt.xlabel("Close Price")
    plt.ylabel("Frequency")
    plt.show()
else:
    print("Column 'close' not found.")
```

```python
# Example 4: Box Plot for Outlier Detection in Close Prices
if 'close' in df.columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df['close'], color='lightgreen')
    plt.title("Box Plot of BTC Close Prices")
    plt.xlabel("Close Price")
    plt.show()
else:
    print("Column 'close' not found.")
```

```python
# Example 5: Correlation Heatmap of Numeric Columns
plt.figure(figsize=(10, 8))
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Open'], label='Open Price')
plt.plot(df['Close'], label='Close Price')
plt.title('Open vs Close Prices')
plt.xlabel('Days')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Volume'], bins=30, color='orange')
plt.title('Volume Distribution')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.show()
```

```python
df['MA_7'] = df['Close'].rolling(window=7).mean()

plt.figure(figsize=(10, 5))
plt.plot(df['Close'], label='Actual Price')
plt.plot(df['MA_7'], label='7-Day MA', linestyle='--')
plt.title('Bitcoin Close Price with 7-Day Moving Average')
plt.xlabel('Days')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(x=df['High'], y=df['Low'], color='purple')
plt.title('High vs Low Prices')
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```
